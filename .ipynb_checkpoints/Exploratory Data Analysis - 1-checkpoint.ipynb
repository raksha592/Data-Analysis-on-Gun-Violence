### In this plot, We see how many guns that were bought had a permit and had been re-checked

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("nics-firearm-background-checks.csv")
```

```python
df['month']=pd.to_datetime(df['month'], format='%Y-%m')
```

```python
permitsGiven=df.groupby("month")["permit","totals"].sum()
```

```python
permitsGiven=pd.DataFrame(permitsGiven)
```

### From the plot below, we see that the total number of guns that were sold is much higher than the permits that were given for these sales. This raises a point of concern

```python
import seaborn as sns
fig, ax = plt.subplots(figsize=(10,10))
ax=sns.lineplot(data=permitsGiven)
plt.ylabel('Total vs Permits')
plt.title('Gun sales over the years')
plt.show()
```

```python
permitsRechecked=df.groupby("month")["permit_recheck","totals"].sum()
```

```python
permitsRechecked=pd.DataFrame(permitsRechecked)
```

### Further more, the number of re-checks that were conducted is much lesser that the total  number of sales. 
### Although, there was an exponential increase in the rechecks between the years 2016-2018

```python
import seaborn as sns
fig, ax = plt.subplots(figsize=(10,10))
ax=sns.lineplot(data=permitsRechecked)
plt.ylabel('Permits re-checks Vs totals')
plt.title('Gun sales over the years')
plt.show()
```
